```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi=[10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi=[20,60,80,85,86,87,70,90,95,90]
```

```python
numpYasListesi=np.array(yasListesi)
numpKiloListesi=np.array(kiloListesi)
```

```python
plt.plot(numpYasListesi,numpKiloListesi,"r")
plt.xlabel("Yas Verisi")
plt.ylabel("Kilo Verisi")
plt.title("Yas ve Kilo GrafiÄŸi")
#(x,y,renk)
```

```python
numpyDizisi=np.linspace(0,10,20)
numpyDizisi
```

```python
numpyDizisi2=numpyDizisi**2
```

```python
plt.plot(numpyDizisi,numpyDizisi2,"g--")
```

```python
plt.subplot(1,2,1)  #1 sÄ±ra, 2 kolon, 1.grafik
plt.plot(numpyDizisi,numpyDizisi2,"r--")
plt.subplot(1,2,2)
plt.plot(numpyDizisi2,numpyDizisi,"b*-")
```

# figure

```python
benimFigure=plt.figure()  #bos kanvas
figureAxes=benimFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpyDizisi,numpyDizisi2,"g")
figureAxes.set_xlabel("x ekseni")
figureAxes.set_ylabel("y ekseni")
figureAxes.set_title("baslik")
```

```python
figure2 = plt.figure()

eksen1=figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2=figure2.add_axes([0.2,0.5,0.2,0.2])

eksen1.plot(numpyDizisi,numpyDizisi2,"g")
eksen1.set_xlabel("X1 ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("Title1")

eksen2.plot(numpyDizisi2,numpyDizisi,"g")
eksen2.set_xlabel("X2 ekseni")
eksen2.set_ylabel("y2 ekseni")
eksen2.set_title("Title2")
```

# subplots

```python
(myFigure, myEksen)=plt.subplots()#colums rows sayÄ±sÄ± ile Ã¶zelleÅŸtir
```

```python
(myFigure, myEksen)=plt.subplots()
myEksen.plot(numpyDizisi,numpyDizisi2,"g")
```

```python
(myFigure, myEksenler)=plt.subplots(nrows=1,ncols=2)

for eksen in myEksenler:
    eksen.plot(numpyDizisi,numpyDizisi2,"g")
    eksen.set_xlabel("x")
plt.tight_layout() # grafik arasÄ± uzaklÄ±k
```

```python
newFigure=plt.figure(dpi=100)        #dpi  Ã§Ã¶zÃ¼nÃ¼rlÃ¼k

yeniEksen=newFigure.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi,numpyDizisi**2,label="Numpy Dizisi **2")
yeniEksen.plot(numpyDizisi,numpyDizisi**3,label="Numpy Dizisi **3")
yeniEksen.legend(loc=0)   #grafik iÃ§indeki tanÄ±mlama =1,2,3,...
```

```python
newFigure.savefig("newFigure.png",dpi=200)
#kaydetme islemi
```
